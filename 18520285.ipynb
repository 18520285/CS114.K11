```
#Khai báo các thư viện cần thiết
import cv2
import numpy as np
import pandas as pd
from google.colab.patches import cv2_imshow
import os
```

```
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt.xml")
```

```
#Cắt lấy khuôn mặt từ một bức ảnh, sau đó chuyển về dạng trắng đen kích cỡ 50*50
def cross(path):
	img = cv2.imread(path)
	img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
	face = face_cascade.detectMultiScale(img)
	print('found', len(face), 'faces', sep=' ')
	if (len(face)==0):
		return None, 0
	else:
		(x, y, w, h) = face[0]
		r = max(w, h) / 2
		centerx = x + w / 2
		centery = y + h / 2
		nx = int(centerx - r)
		ny = int(centery - r)
		nr = int(r * 2)
		faceimg = img[ny:ny+nr, nx:nx+nr]
		faceimg = cv2.resize(faceimg,(50,50))
	return faceimg, len(face)
```

```
#Tạo tập training từ data
s_path = '/content/drive/My Drive/train'
count = 1
number_face = 0
dectec_t = 0
k = 0
X = pd.DataFrame(columns=range(2500))
Y = pd.DataFrame(columns=['label'])
for i in os.listdir(s_path):
    print('==== process image', i, ':', count, 'count of', len(os.listdir(s_path)), sep=' ')
    face, k =cross(s_path + "/" + i)
    number_face += k
    if face is not None:
      flat = face.flatten()
      t = pd.DataFrame([flat])
      dectec_t += 1
    else:
      print('Can\'t find face')
      t = pd.DataFrame([[0]*2500], columns=range(2500))
    X=X.append(t, ignore_index=True)
    Y=Y.append({'label':i[0:3]},ignore_index=True)
    count = count+1
print('Face detection/face real:',number_face,'/',len(os.listdir(s_path)))
print('Accuraccy dectect face:', dectec_t/len(os.listdir(s_path)))
```

```
# chọn thuật toán học
from sklearn.svm import SVC
model = SVC(gamma='auto')

model.fit(X, Y.values.ravel())

print(model)
```
